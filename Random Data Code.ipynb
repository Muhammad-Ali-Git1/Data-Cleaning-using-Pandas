```python
import pandas as pd
import numpy as np
df=pd.read_csv('New dataset.csv')
```

```python
df=df.set_index(df['ID'])
```

```python
df=df.drop(columns="ID")
```

```python
name_dic={
    "Alee":"Ali",
    "Jon":"John",
    "Mari":"Mary",
    "Usmn":"Usman",
    "Frah":"Farah",
    "Imrn":"Imran",
    "Aysha":"Ayesha",
    "Kmrn":"Kamran",
    "Zra":"Zara",
}    
df.loc[:,'Name']=df['Name'].replace(name_dic)
```

```python
df=df.dropna()
```

```python
Age_dic={
    "Twenty":"20",
    "twenty-six":"26",
    "Forty":"40",
}
df['Age']=df['Age'].replace(Age_dic)
```

```python
df['JoinDate']=pd.to_datetime(df['JoinDate'],format="mixed")
```

```python
Salary_dic={
    "50k":"50000",
    "48k":"48000",
    "51k":"51000",
    "58k":"58000",
    "70k":"70000",
    "65k":"65000",
    "55k":"55000",
    "53k":"53000"
}
df['Salary']=df['Salary'].replace(Salary_dic)
```

```python
df['Salary']=df['Salary'].apply(lambda x:"{:,}".format(int(x)))
```

```python
df['Dept']=df['Dept'].str.title()
```

```python
Dept_dic={
    "It":"IT",
    "Hr":"HR",
}
df['Dept']=df['Dept'].replace(Dept_dic)
```

```python
df['Phone']=df['Phone'].astype(str)
```

```python
df['Phone']=df['Phone'].str.strip().replace(r"[^0-9]","",regex=True)
```

```python
df['Phone']=df['Phone'].replace(r"^92","0",regex=True)
```

```python
df["Phone"] = df["Phone"].apply(lambda x: x[:11] if len(x) > 11 else x.zfill(11))
```

```python
df["Email"] = df["Email"].apply(lambda x: x if x.endswith(".com") else x + ".com")
```

```python
Remarks_dic={
    "gd":"Good",
    "gud":"Good",
    "avg":"Average",
    "Exclnt":"Excellent",
}
df['Remarks']=df['Remarks'].replace(Remarks_dic)
```
