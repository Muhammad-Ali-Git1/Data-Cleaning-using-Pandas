```python
import pandas as pd
import numpy as np
df=pd.read_csv('Messy Company Data.csv')
```

```python
df
```

```python
df=df.sort_values(by='Employee_ID')
```

```python
df=df.reset_index()
```

```python
df
```

```python
df=df.drop(columns='index')
```

```python
df=df.drop_duplicates()
```

```python
df.columns=df.columns.str.strip()
```

```python
df=df.dropna(subset='Name')
```

```python
df['Join_Date']
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='mixed', errors='coerce')
```

```python
df['Join_Date'].dtype
```

```python

df['Join_Date']=df['Join_Date'].dt.date
```

```python
df['Join_Date']=df['Join_Date'].fillna('Not Provided')
```

```python
df['Salary']=df['Salary'].astype(str)
```

```python
df['Salary']=df['Salary'].str.strip()
```

```python
df['Salary']=df['Salary'].str.replace(r'[^0-9]','',regex=True)
```

```python
df['Salary']=df['Salary'].fillna('None')
```

```python
df['Salary']=pd.to_numeric(df['Salary'])
```

```python
df['Salary']=df['Salary'].astype(int)
```

```python
df['Salary']=df['Salary'].apply(lambda x: '{:,}'.format(int(x)))
```

```python
df['Bonus']=df['Bonus'].replace('five thousand',5000)
```

```python
df['Remote_Work']=df['Remote_Work'].replace({'true':'Yes','TRUE':'Yes','false':'No','FALSE':'No','1':'Yes','0':'No'})
```

```python
df['Phone']=df['Phone'].replace('invalid_number',np.nan)
```

```python
df['Phone']=df['Phone'].replace(np.nan,'1')
```

```python
df['Phone']=df['Phone'].replace(r'[^0-9]','',regex=True)
```

```python
df['Phone']=df['Phone'].str[-11:]
```

```python
df['Phone']=df['Phone'].replace('1',np.nan)
```

```python
df['Phone']=df['Phone'].fillna('Not Provided')
```

```python
df['Performance_Score']=df['Performance_Score'].replace({'Four':'4','four':'4'})
```

```python
df['Performance_Score']=df['Performance_Score'].astype(float)
```

```python
df['Performance_Score']=df['Performance_Score'].fillna(df['Performance_Score'].mean())
```

```python
df['Performance_Score']=df['Performance_Score'].round(2)
```

```python
df['Last_Promotion_Date']=df['Last_Promotion_Date'].fillna('Not Provided')
```

```python
df['Last_Promotion_Date']=df['Last_Promotion_Date'].astype(str)
```

```python
df['Last_Promotion_Date']=pd.to_datetime(df['Last_Promotion_Date'],format='mixed',errors='coerce')
```

```python
df['Last_Promotion_Date']=df['Last_Promotion_Date'].dt.date
```

```python
df['Last_Promotion_Date']=df['Last_Promotion_Date'].fillna('Not Provided')
```

```python
df
```

```python
df.to_csv('Cleaned Company Dataset.csv')
```
