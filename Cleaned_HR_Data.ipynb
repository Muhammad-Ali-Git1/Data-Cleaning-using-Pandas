```python
import pandas as pd
import numpy as np
df=pd.read_csv('messy_HR_data.csv')
```

```python
df
```

```python
df['Name']=df['Name'].str.title()
```

```python
df
```

```python
df['Age']=df['Age'].replace('thirty',30)
```

```python
df
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df
```

```python
df['Salary']=df['Salary'].replace('SIXTY THOUSAND',60000)
```

```python
df['Salary']=df['Salary'].str.strip().str.upper()
```

```python
df['Salary']=df['Salary'].replace(['NaN','NAN'],np.nan)
```

```python
df['Salary']=pd.to_numeric(df['Salary'])
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

```python
df.rename(columns={'Joining Date':'JDate'}, inplace=True)
```

```python
df['Salary']=df['Salary'].astype(int)
```

```python
df['Salary']=df['Salary'].apply(lambda x: "{:,}".format(int(x)))
```

```python
df['Salary']=df['Salary'].astype(str)+'$'
```

```python
df['JDate'] = df['JDate'].str.replace('April', '04')
```

```python
df['JDate'] = pd.to_datetime(df['JDate'], errors='coerce', format='mixed')
```

```python
df
```

```python
df
```

```python
df['Gender']=df['Gender'].replace({'Female':'F','Male':'M'})
```

```python
df
```

```python
df['Phone Number']=df['Phone Number'].astype(str)
```

```python
df['Phone Number']=df['Phone Number'].str.replace(r'\D','',regex=True)
```

```python
df['Phone Number']=df['Phone Number'].str.replace(r'(\d{3})(\d{3})(\d{3})',r'\1-\2-\3',regex=True)
```

```python
df
```

```python
df['Email']=df['Email'].fillna('No Email Provided')
```

```python
df
```

```python
df.rename(columns={'JDate':'Joining Date'},inplace=True)
```

```python
df
```

```python
df.to_csv('Cleaned HR Data.csv')
```
